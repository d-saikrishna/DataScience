```python
sns.kdeplot(data)
plt.show()
```

```python
means = []
sample_size = 100
# Lets calculate 10000 means
for i in range(20000):
    # Pick 3 elements randomly from the array
    sample = random.sample(list(data), k=sample_size)
    mean = np.array(sample).mean()
    means.append(mean)
```

```python
len(means)
```

```python
sns.kdeplot(means, label='Sampling Distribution')
# Overlay normal curve
mu, std = np.mean(means), np.std(means)
x = np.random.normal(loc=mu, scale=std, size=10000)
sns.kdeplot(x, c='r', label='Normal Distribution')
plt.xlabel("Means")
plt.legend()
#plt.plot(x, norm.pdf(x, mu, std), 'r-', lw=2)
plt.show()
```
