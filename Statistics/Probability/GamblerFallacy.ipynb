```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_simulations = 10000000
tosses = []
for _ in range(n_simulations):
    toss = np.random.binomial(1, 0.5,1)[0]
    tosses.append(toss)
```

```python
means = []
for i in range(len(tosses)-10):
    sequence = np.array(tosses[i:i+10])
    if sequence.sum() == 10:
        next_sequence = np.array(tosses[i+10:i+20])
        means.append(next_sequence.mean())
```

```python
np.array(means).mean()
```

```python
k = []
for i in range(len(means)):
    k.append(np.array(means[:i]).mean())
```

```python
plt.plot(range(len(means)), k)
plt.show()
```
