Suppose we play a game where we start with c dollars. On each play of the game you either double or halve your money, with equal probability.
What is your expected fortune after n trials

```python
wealth = 100
trial = 0
num_trials = 10
while trial < num_trials:
    wealth = (0.5*2*wealth - 0.5*wealth/2)
    trial = trial+1

print('After {} trials, fortune left is: {}'.format(num_trials, wealth))
```

Clearly, you should not make such bets

**Q. A bomb has 50% probability of correctly hitting the target. 2 Bomb hits are required to eliminate the target. How many bombs are required to be thrown to eliminate the target with 99% probability?**

A: Binomial distribution. `p = 0.5`; `q=0.5`
`n = number of bombs`
`X = number of bomb hits`

Expected Value = np.

It is a mistake to just equate Expected value to 2 bomb hits. By that we will get that `np=2` ==> `n=4` -- 4 bombs are enough to get 2 bomb hits. But is this at 99% probability as the question asked?

What is the probability of of getting two bomb hits when n=4?
P(X=2), n=4 ==> 4C2*(1/2)^4 = 6/16 = 0.375

Atleast 2? P(X=2)+P(X=3)+P(X=4) = 0.6875 < 99%

Therefore with 4 bombs only 68.75% chance of completely eliminating the target.

To solve this problem, one must understand the Probability distribution well.

```python
# Probability Mass Function
n=4
import math

Ps = []
for x in range(0,5):
    #Probability of getting x hits
    P_xhits = math.comb(n, x) * (1/2**n)
    Ps.append(P_xhits)

import matplotlib.pyplot as plt
import numpy as np

plt.step(range(0,5),Ps,'g',label='PMF') 
plt.step(range(0,5),np.cumsum(Ps), 'r', label='CDF')
plt.xlabel('Number of Bomb Hits')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

```python
# we have to find n, for which this peak should occur at 99%
Ps = []

for n in range(2,20):
    #Probability of getting atleast 2 hits = 1 - Probability of 0 hits - Probability of 1 hit
    P_2hits = 1 - math.comb(n, 0) * (1/2**n) - math.comb(n, 1) * (1/2**n)
    Ps.append(P_2hits)

import matplotlib.pyplot as plt

plt.plot(range(2,20),Ps,'g') 
plt.xlabel('Number of Bombs')
plt.ylabel('Probability')
plt.axhline(0.99)
plt.show()
#n= 11 - >99% Probability
```
